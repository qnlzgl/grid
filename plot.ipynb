```python
import pandas as pd

```

```python
import matplotlib.pyplot as plt

def plot_predictions(data):
    """
    Plots the actual values and predicted values with error bars.

    Parameters:
    data (pd.DataFrame): The dataframe containing the data to plot.
    """
    plt.figure(figsize=(20,6))

    # Plot the actual test values
    plt.plot(data['time_stamp'], data['y_test'], label='Actual (y_test)', color='black', linewidth=2)

    # Plot the machine learning predictions with shaded error areas
    plt.plot(data['time_stamp'], data['y_ml_pred_mean'], label='ML Prediction (y_ml_pred_mean)', color='blue', alpha=0.7)
    plt.fill_between(data['time_stamp'], 
                     data['y_ml_pred_mean'] - data['y_ml_pred_std'], 
                     data['y_ml_pred_mean'] + data['y_ml_pred_std'], 
                     color='blue', alpha=0.3)

    # Plot the data processing predictions with shaded error areas
    plt.plot(data['time_stamp'], data['y_dp_pred_mean'], label='DP Prediction (y_dp_pred_mean)', color='green', alpha=0.7)
    plt.fill_between(data['time_stamp'], 
                     data['y_dp_pred_mean'] - data['y_dp_pred_std'], 
                     data['y_dp_pred_mean'] + data['y_dp_pred_std'], 
                     color='green', alpha=0.3)

    # Add titles and labels
    plt.title('Actual vs Predicted Values with Error Bars', fontsize=14)
    plt.xlabel('Time Stamp', fontsize=12)
    plt.ylabel('Values', fontsize=12)

    # Show the legend
    plt.legend()

    # Show the plot
    plt.tight_layout()
    plt.show()

# Example usage

# List of file paths
file_paths = [
    '/home/qnlzgl/sequence-model-and-gaussian-process-for-frequency-prediction-main/data/prediction_example_0_0.csv',
    '/home/qnlzgl/sequence-model-and-gaussian-process-for-frequency-prediction-main/data/prediction_example_0_1.csv',
    '/home/qnlzgl/sequence-model-and-gaussian-process-for-frequency-prediction-main/data/prediction_example_1_0.csv',
    '/home/qnlzgl/sequence-model-and-gaussian-process-for-frequency-prediction-main/data/prediction_example_1_1.csv'
]

# Plot predictions for each file
for file_path in file_paths:
    data = pd.read_csv(file_path)
    plot_predictions(data)

```
